2. Encontrar o Maior NÃºmero

```python
# Encontre o maior nÃºmero em uma lista
numeros = [34, 12, 89, 5, 67, 23]


# Seu cÃ³digo aqui
maior = max(numeros)
print(f"Maior nÃºmero: {maior}")  # Deve imprimir 89
```

3. Adicionar Elementos com CondiÃ§Ã£o

```python
# Dada uma lista de nÃºmeros, crie uma nova lista apenas com nÃºmeros maiores que 50
numeros = [25, 60, 15, 80, 45, 90, 30]
maiores_que_50 = []

# Seu cÃ³digo aqui (use append com if)
for n in numeros:
    if n > 50:
        maiores_que_50.append(n)

print(maiores_que_50)  # Deve imprimir: [60, 80, 90]
```

5. Controle de Gastos Pessoais

```python
print("=== CONTROLE DE GASTOS PESSOAIS ===")

# Listas para armazenar os dados
nomes_gastos = []
valores_gastos = []

# Pergunta a quantidade de gastos
qtd = int(input("Quantos gastos deseja registrar? "))

# Loop para coletar os gastos
for i in range(qtd):
    nome = input(f"Nome do gasto {i+1}: ")
    valor = float(input(f"Valor do gasto {i+1}: R$ "))
    
    # Adiciona Ã s listas usando APPEND
    nomes_gastos.append(nome)
    valores_gastos.append(valor)

# Calcula o total dos gastos
total_gastos = sum(valores_gastos)

# Exibe os resultados
print(f"\nTOTAL GASTO: R$ {total_gastos:.2f}")

# Verifica se estÃ¡ gastando muito
if total_gastos > 500:
    print("âš ï¸ VocÃª estÃ¡ gastando como um dragÃ£o faminto! Cuidado! ğŸ‰")
else:
    print("Tudo sob controle. Seu bolso agradece ğŸŒ¿")
```
